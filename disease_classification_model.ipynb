```python
import math

def has_nan(lst):
    for x in lst:
        if isinstance(x, float) and math.isnan(x):
            return True
    return False


if has_nan(unique_list):
    print("The list contains NaN values.")
else:
    print("The list does not contain NaN values.")
```

```python
def remove_nan(lst):
    return [x for x in lst if not (isinstance(x, float) and math.isnan(x))]

cleaned_list = remove_nan(unique_list)
```

```python
from sklearn.model_selection import train_test_split

y=df['disease'] 
X=df.drop('disease',axis=1) 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier(n_estimators=300,criterion='gini',random_state=42)
```

```python
rfc.fit(X_train,y_train)
```

```python
prediction=rfc.predict(X_test)
```

```python
from sklearn import metrics
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, prediction))
```

```python
importances=rfc.feature_importances_
```

```python
num_features_to_select = 56  # Adjust this based on your needs
important_feature_indices = importances.argsort()[-num_features_to_select:][::-1]
selected_features = X_train.columns[important_feature_indices]
```

```python
X_train_new = X_train[selected_features]
X_test_new = X_test[selected_features]
```

```python
rf_classifier_new = RandomForestClassifier(n_estimators=100, random_state=42)
rf_classifier_new.fit(X_train_new, y_train)
```

```python
y_pred_new = rf_classifier_new.predict(X_test_new)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test,y_pred_new ))
```
